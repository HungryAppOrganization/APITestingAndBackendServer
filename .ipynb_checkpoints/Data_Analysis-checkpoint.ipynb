Now, we are going to show each item and the # of swipes per item.
To do this, I will aggregate all the swipes together, then condense them down. The idea is to have a list. Item : #swipes, #left, #right

```python
username = "Rachel s iPhone"
def getUserSwipes(username):
    swipes = myDb.db.child("users").order_by_child("username").equal_to(username).get().val().values()[0]['swipes'].values()
    return swipes
```

```python
allUsers = myDb.getAllUsers()
allSwipes = []
for user in allUsers:
    allSwipes = allSwipes + getUserSwipes(user)
print("---------- Swipes ----------")
print(allSwipes)
#Now this may be slightly hard. I need to compress this dict. 
```

```python
def getDictSwipes(userSwip):
    myDict = {}
    #print('----- Swipe ----')
    for sw in userSwip:
        myDict[sw['cardID']] = sw['swipe']
    #print(sw)
    #print(sw['swipe'],sw['cardID'])
    #myDict[sw['cardID']]=sw['swipe']
    #print(myDict)
    return myDict
```

```python
print(getDictSwipes(getUserSwipes(user)))
```

```python
allMyDicts = []
allUsers = myDb.getAllUsers()
for user in allUsers:
    allMyDicts.append(getDictSwipes(getUserSwipes(user)))
print("---------- Swipes ----------")
print(allMyDicts)
```

```python
from collections import defaultdict
dd = defaultdict(list)
for d in allMyDicts: # you can list as many input dicts as you want here
    for key, value in d.iteritems():
        dd[key].append(value)
#print(dd)
finalSwipesPerItem = dict(dd)
print(finalSwipesPerItem)
#Print this to a file
```

```python
itemSwipes = "item_swipes_5_21.csv"
outFile = open(itemSwipes,'w')
for key, val in finalSwipesPerItem.items():
    outFile.write(str(key) + "," + ",".join([str(x) for x in val]) +"\n")
outFile.flush()
```
